```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_curve, auc
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import MinMaxScaler, Binarizer
```

```python
df = pd.read_csv('creditcard.csv')
print(df.head())

print(df.info())
print('\n')

print(df.describe())
```

```python
# Detailed class distribution before SMOTE
plt.figure(figsize=(12, 6))
sns.countplot(x='Class', data=df)
plt.title('Class Distribution Before SMOTE')
plt.xlabel('Class')
plt.ylabel('Count')

# Annotate counts on the bars
for p in plt.gca().patches:
    plt.gca().annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                       ha='center', va='center', xytext=(0, 9), textcoords='offset points')

plt.show()

# Pie chart for class distribution before SMOTE
class_counts = df['Class'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', colors=['skyblue', 'salmon'], startangle=140)
plt.title('Class Distribution Before SMOTE (Pie Chart)')
plt.show()
```

```python
# Feature and target selection
X = df.drop('Class', axis=1)  # Features
y = df['Class']               # Target variable (fraud or not fraud)

# Preprocessing: Normalize the feature data
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Binarize the feature data for BernoulliNB
binarizer = Binarizer()
X_binarized = binarizer.fit_transform(X_scaled)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_binarized, y, test_size=0.2, random_state=42)

# Apply SMOTE to oversample the minority class in the training set
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_train, y_train)

# Detailed class distribution after SMOTE
plt.figure(figsize=(12, 6))
sns.countplot(x=y_res)
plt.title('Class Distribution After SMOTE')
plt.xlabel('Class')
plt.ylabel('Count')

# Annotate counts on the bars
for p in plt.gca().patches:
    plt.gca().annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                       ha='center', va='center', xytext=(0, 9), textcoords='offset points')

plt.show()

# Pie chart for class distribution after SMOTE
class_counts_res = pd.Series(y_res).value_counts()
plt.figure(figsize=(8, 8))
plt.pie(class_counts_res, labels=class_counts_res.index, autopct='%1.1f%%', colors=['skyblue', 'salmon'], startangle=140)
plt.title('Class Distribution After SMOTE (Pie Chart)')
plt.show()
```

```python
# Train the BernoulliNB model on the resampled dataset
bnb = BernoulliNB()
bnb.fit(X_res, y_res)
```

```python
# Make predictions on the test set
y_pred = bnb.predict(X_test)

# Evaluate the model
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nAccuracy Score:")
print(accuracy_score(y_test, y_pred))

# Visualize the Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.ylabel('Actual Class')
plt.xlabel('Predicted Class')
plt.show()

# ROC Curve and AUC
y_pred_proba = bnb.predict_proba(X_test)[:, 1]
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.show()
```
